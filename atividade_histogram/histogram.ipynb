```python
import numpy as np
from skimage.io import imread,imsave
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.exposure import histogram
from skimage.exposure import equalize_hist
```

```python
img = imread('yuyu-hakusho.jpg')
imgGray = rgb2gray(img)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax1.set_title('Imagem Original')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(imgGray,cmap='gray')
ax2.set_title('Imagem em tons de cinza')
```

```python
hist = histogram(imgGray)
hist
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(imgGray,cmap='gray')
ax1.set_title('Imagem em tons de cinza')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(imgGray.ravel(),range(256))
ax2.set_title('Histograma')
```
